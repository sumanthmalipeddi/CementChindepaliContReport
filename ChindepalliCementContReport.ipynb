```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
contractor_df = pd.read_excel("filepath",skiprows=1)

cont_ben = contractor_df[["Contract_name","Beneficiary Id"]]

cement_df = pd.read_excel("filepath",skiprows=1)

cement_filter = cement_df[["Layout / Own sites","Layout ID / Layout Name","Beneficiary ID /  Beneficiary Name /  Father/Husband Name",
                               "Stage","Issued Quantity/ Material issued value"]]

cement_chindepalli = cement_filter[cement_filter["Layout / Own sites"] == "010841LA111149"]

# Split the 'EA/WAS ID / EA/WAS Name / EA/WAS Mobile No.' column by '/'

cement_chindepalli[['Beneficiary ID', 'Beneficiary Name', 'Father/Husband Name']] = cement_chindepalli["Beneficiary ID /  Beneficiary Name /  Father/Husband Name"].str.split('/', expand=True)

cement_chindepalli[['Issued Quantity', 'Material issued value']] = cement_chindepalli["Issued Quantity/ Material issued value"].str.split('/', expand=True)

#df['EventCount'] = df['Event'].str.split("/").str.len()0

cement_chindepalli["Ben_16"] = cement_chindepalli["Beneficiary ID"].str.strip()

result = pd.merge(cement_chindepalli, cont_ben, how='left', left_on='Ben_16', right_on='Beneficiary Id')

result["Issued Quantity"] = result["Issued Quantity"].astype('float')

trial = result

fin = result.groupby(["Contract_name"])["Issued Quantity"].sum()

#print(fin)

used = fin

used = used.to_frame()

print(used)
```

```python
phyqnty = []

for i in used.index:
    
    phyqnty.append(int(input(f'The physical quantity for {i}:')))

used['Physical Quantity'] = phyqnty  

used['Difference'] = used['Physical Quantity'] - used['Issued Quantity']
```

```python
# Increase the figure size
plt.figure(figsize=(20, 20))

# Add labels to the plot
used.plot.bar()
plt.xlabel("Contract Name")
plt.ylabel("Quantity")
plt.title("Difference between Issued and Physical Quantity for Contracts")



# Save the plot as a PNG image
plt.savefig("difference_plot.png")

# Display the plot
plt.show()
```
